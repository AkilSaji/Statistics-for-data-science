```python
import numpy as np
import pandas as pd
```

```python
beml=pd.read_csv("/content/BEML.csv")
beml
```

```python
glaxo=pd.read_csv("/content/GLAXO .csv")
glaxo
```

```python
glaxo.head(5)
```

```python
beml=beml[["Date","Close"]]
glaxo=glaxo[["Date","Close"]]
beml
```

```python
glaxo
```

```python
glaxo=glaxo.set_index(pd.DatetimeIndex(glaxo["Date"]))
beml=beml.set_index(pd.DatetimeIndex(beml["Date"]))
beml
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
plt.plot(beml.Close)
plt.xlabel("Time")
plt.ylabel("Close Price")
```

```python
plt.plot(glaxo.Close)
plt.xlabel("Time")
plt.ylabel("Close Price")
```

```python
glaxo["gain"]=glaxo.Close.pct_change(periods=1)
beml["gain"]=beml.Close.pct_change(periods=1)
```

```python
glaxo=glaxo.dropna()
beml=beml.dropna()
```

```python
plt.figure(figsize=(8,6));
plt.plot(glaxo.index,glaxo.gain);
plt.plot(glaxo.Close)
plt.xlabel("Time")
plt.ylabel("Close Price")
plt.ylabel("gain");
```

```python
sn.distplot(glaxo.gain,label="Glaxo");
plt.xlabel("gain")
plt.ylabel("Density")
plt.legend()
```

```python
sn.distplot(glaxo.gain,label="beml");
plt.xlabel("gain")
plt.ylabel("Density")
plt.legend()
```

```python
print("Mean",round(glaxo.gain.mean(),4))
print("Standar Deviation",round(glaxo.gain.std(),4))
```

```python
print("Mean",round(beml.gain.mean(),4))
print("Standar Deviation",round(beml.gain.std(),4))
```
